```python
import pandas as pd

# Load datasets
fake = pd.read_csv("Fake.csv")
true = pd.read_csv("True.csv")

# Create labels
fake["label"] = 0
true["label"] = 1

# Keep only headlines
fake = fake[["title","label"]]
true = true[["title","label"]]

# Take equal samples (7000 each)
fake = fake.sample(n=7000, random_state=42)
true = true.sample(n=7000, random_state=42)

print("Fake count:", len(fake))
print("Real count:", len(true))

# Combine and shuffle
data = pd.concat([fake, true])
data = data.sample(frac=1, random_state=42)

# Rename column for ML
data = data.rename(columns={"title":"text"})

print("Total dataset:", len(data))

# Show sample
data.head()


```

```python
import string

def clean_text(text):
    text = str(text).lower()                  # lowercase
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    return text

data["text"] = data["text"].apply(clean_text)

data.head()
```

```python
data = data.sample(frac=0.05, random_state=42)
print("Dataset size:", len(data))

```

```python
from sklearn.model_selection import train_test_split

X = data["text"]
y = data["label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42
)


```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=100)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

```

```python
X_train_dense = X_train_vec.toarray()
X_test_dense = X_test_vec.toarray()

```

```python
from sklearn.svm import LinearSVC

model = LinearSVC(class_weight='balanced')
model.fit(X_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test_vec)
print("Accuracy:", accuracy_score(y_test, y_pred))


```

```python
def predict_news(text):
    text = clean_text(text)
    vec = vectorizer.transform([text])
    pred = model.predict(vec)[0]

    if pred == 1:
        print("REAL NEWS")
    else:
        print("FAKE NEWS")

predict_news("Government announces new education policy")
predict_news("Scientists hide miracle cure discovered in secret lab")
predict_news("Secret organization controlling world leaders exposed by insider")
predict_news("Man claims he teleported to another planet and brought back photos")


predict_news("Ancient pyramid found on the moon according to leaked NASA photos")





```
